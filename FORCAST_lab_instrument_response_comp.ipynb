# Comparison to the previous cooldown (Instrument Response)

## Import needed modules

```python
import pandas as pd
import seaborn as sns
sns.set(style="ticks", color_codes=True)
%matplotlib inline
```

## Parameters to change

Here define the files to be used and the path and name for your plot:
* **comp_file**: The previous cooldown's result file. If you don't have this, it is in confluence in the previous cooldown results page. If for some reason this was not created by FORCAST_lab_bg_photonrate.ipynb, make sure that the columns match
* **cd_file**: The results for the current cooldown. You can directly use the file created by FORCAST_lab_bg_photonrate.ipynb 
* **plot_name**: The path and name for the result plot


## THIS IS THE ONLY PLACE YOU SHOULD NEED TO CHANGE ANYTHING

```python
comp_file='Cooldown_results/Cooldown_76/FORCAST_bg_LB076_20191003_final_v2.xlsx'
cd_file='Cooldown_results/Cooldown_77/FORCAST_bg_LB077_20191120.xlsx'
plot_name='Cooldown_results/Cooldown_77/LB077_comp_to_previous_IR.png'
```

## Read in the excel files

```python
comp=pd.read_excel(comp_file)
cool=pd.read_excel(cd_file)
```

## Add the information to a new dataframe
Here we create a new dataframe for the data to be plotted. The column names are also changed so that the merged dataframe is easier to manipulate

```python
df_cd = cool.copy(deep=False)
df_cd = df_cd.drop(columns=['filename','ilowcap','aperture','dichroic','framerate','detchan','detbias','eperadu','raw_siglevel','pa_offset','full_pe_rate'])
df_comp = comp.copy(deep=False)
df_comp = df_comp.drop(columns=['filename','wavelength','ilowcap','aperture','dichroic','framerate','detchan','detbias','eperadu','raw_siglevel','pa_offset','full_pe_rate'])
df_cd=df_cd.rename(columns={'pe_rate':'cd_pe_rate'})
df_comp=df_comp.rename(columns={'pe_rate':'comp_pe_rate'})
```

```python
df = pd.merge(left=df_cd, right=df_comp, on='instcfgs', how='left')
df
```

## Plot the figure with old and new data

Finally we create a plot with seaborn amd the plot is then saved as a png. The current data are the dots and the previous data are the X.

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_style('whitegrid')
sns_plot = sns.scatterplot(x='wavelength', y='cd_pe_rate', data=df, hue='instcfgs', marker='o', legend=False, s=100)
sns_plot = sns.scatterplot(x='wavelength', y='comp_pe_rate', data=df, hue='instcfgs', marker='x', legend=False, s=100)
#sns_plot.set(xlabel = 'Wavelength [$\mu$m]', ylabel = 'pe rate [$Me^-$/sec]')
sns_plot.set_xlabel('Wavelength [$\mu$m]',fontsize=20)
sns_plot.set_ylabel('pe rate [$Me^-$/sec]',fontsize=20)
```

```python
fig = sns_plot.get_figure()
fig.savefig(plot_name, bbox_inches='tight')
```
